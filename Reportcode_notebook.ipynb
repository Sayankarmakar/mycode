```python
import os
import math
# import required libraries for dataframe and visualization

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# import required libraries for clustering

import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree

# Reading the data on which analysis needs to be done

retail = pd.read_csv('../input/online-retail-ii-uci/online_retail_II.csv',)
retail.head()
```

```python
retail.shape
```

```python
retail.info()


retail.describe()
```

```python
sqi= retail[["StockCode","Quantity","InvoiceDate","Price"]]
sqi.head()
```

```python
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'],format='%Y-%m-%d %H:%M')
```

```python
date1= pd.to_datetime('now')
retail["diff"]= date1 - retail["InvoiceDate"]
retail.shape
```

```python
sqi= retail[["StockCode","Quantity","InvoiceDate","Price","diff"]]
sqi['diff']= sqi['diff'].dt.days
sqi.head()

```

```python
df1= pd.DataFrame(sqi, columns=['StockCode','Quantity','InvoiceDate','diff','Price'])
df1= df1.sample(n=100)
print(df1)
```

```python
df1.plot(kind='scatter',x='Price',y='Quantity')
plt.xlim(0,20)
plt.ylim(0,40)
plt.show()

```

```python
df2=df1[['Price','diff']]

print(df2)
```

```python

km=KMeans(n_clusters=5, max_iter=100)
km.fit(df2)
```

```python
km.labels_
```

```python
 #Elbow-curve/SSD

ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]
for num_clusters in range_n_clusters:
    km= KMeans(n_clusters=num_clusters, max_iter=100)
    km.fit(df2)
    
    ssd.append(km.inertia_)
    
# plot the SSDs for each n_clusters
plt.plot(ssd)
```

```python
# Silhouette analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    km = KMeans(n_clusters=num_clusters, max_iter=50)
    km.fit(df2)
    
    cluster_labels = km.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(df2, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
#for 7 clusters
km = KMeans(n_clusters=7, max_iter=100)
km.fit(df2)
```

```python
km.labels_
```

```python
# assign the label
df2['Cluster_Id'] = km.labels_
df2.head()
```

```python
sns.boxplot(x='Cluster_Id', y='Price', data=df2)
```
